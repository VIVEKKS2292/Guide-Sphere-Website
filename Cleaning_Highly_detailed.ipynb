```python
import pandas as pd
```

```python
df = pd.read_csv('Highly_detailed.csv')
```

```python
df = df[['course_url', 'course_img', 'course_duration', 'course_price',
       'course_price_original', 'course_instrutor', 'course_title',
       'course_short_desc', 'course_rating', 'course_date', 'course_language',
       'course_outcome_text', 'course_outcome_html', 'course_about_text',
       'course_about_html','course_about_text2',
       'course_about_html_2']]
```

```python
df.head(50)
```

```python
df.columns
```

```python
def longer_course_about_text(row):
    course_about_text = str(row['course_about_text'])  # Convert to string
    course_about_text2 = str(row['course_about_text2'])  # Convert to string
    if len(course_about_text) >= len(course_about_text2):
        return course_about_text
    else:
        return course_about_text2
```

```python
df['course_about_text'] = df.apply(longer_course_about_text, axis=1)
```

```python
def longer_course_about_html(row):
    course_about_html = str(row['course_about_html'])  # Convert to string
    course_about_html_2 = str(row['course_about_html_2'])  # Convert to string
    if len(course_about_html) >= len(course_about_html_2):
        return course_about_html
    else:
        return course_about_html_2
```

```python
df['course_about_html'] = df.apply(longer_course_about_html, axis=1)
```

```python
df = df[['course_url', 'course_img', 'course_duration', 'course_price',
       'course_price_original', 'course_instrutor', 'course_title',
       'course_short_desc', 'course_rating', 'course_date', 'course_language',
       'course_outcome_text', 'course_outcome_html', 'course_about_text',
       'course_about_html']]
```

```python
len(df)
```

```python
course_ids = ['c' + str(i).zfill(5) for i in range(1, len(df) + 1)]
```

```python
df['course_id'] = course_ids
```

```python
df.to_csv('modified_Highly_detailed_file.csv', index=False)
```

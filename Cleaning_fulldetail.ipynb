```python
import pandas as pd
```

```python
df = pd.read_csv('fulldetail.csv')
```

```python
df
```

```python
def longer_description(row):
    long_desc = str(row['longDesc'])  # Convert to string
    text = str(row['Text'])  # Convert to string
    if len(long_desc) >= len(text):
        return long_desc
    else:
        return text
```

```python
df['Long Description'] = df.apply(longer_description, axis=1)
```

```python
df = df[['url','img','length','Title','shortDesc','creator','rating','language','date','Long Description',]]
```

```python
df.to_csv('modified_fulldetail_file.csv', index=False)
```
